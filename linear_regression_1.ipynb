```python
x1 = df.drop(["Sales", "total_spent"], axis=1)
x1.head()
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly_conv = PolynomialFeatures(degree=2, include_bias=False)
```

```python
poly_conv.fit(x1)
```

```python
poly_features = poly_conv.transform(x)
```

```python
poly_features.shape
```

```python
x1.loc[0]
```

```python
poly_features[0]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(poly_features, y, test_size = 0.33, random_state = 41)
```

```python
model1 = LinearRegression()
```

```python
model1.fit(X_train, y_train)
```

```python
test_pred = model1.predict(X_test)
test_pred
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
MAE = mean_absolute_error(y_test, test_pred)
MAE
```

```python
MSE = mean_squared_error(y_test, test_pred)
MSE
```

```python
RMSE = np.sqrt(MSE)
RMSE
```

```python
model1.coef_
```

```python
train_rmse_errors = []
test_rmse_errors = []

for d in range(1, 10):
    poly_converter = PolynomialFeatures(degree=d, include_bias=False)
    poly_features = poly_converter.fit_transform(x1)
    X_train, X_test, y_train, y_test = train_test_split(poly_features, y, test_size = 0.33, random_state = 41)

    model2 = LinearRegression()
    model2.fit(X_train, y_train)

    train_pred = model2.predict(X_train)
    test_pred = model2.predict(X_test)

    train_rmse = np.sqrt(mean_squared_error(y_train, train_pred))
    test_rmse = np.sqrt(mean_squared_error(y_test, test_pred ))

    train_rmse_errors.append(train_rmse)
    test_rmse_errors.append(test_rmse)
```

```python
train_rmse_errors
```

```python
test_rmse_errors #Overfitting after 5th degree
```

```python
plt.plot(range(1, 6), train_rmse_errors[:5], label = "TRAIN_RMSE")
plt.plot(range(1, 6), test_rmse_errors[:5], label = "TEST_RMSE")
plt.xlabel("MODEL COMPLEXITY")
plt.ylabel("ERRORS")
plt.legend()
```

```python
plt.plot(range(1, 10), train_rmse_errors, label = "TRAIN_RMSE")
plt.plot(range(1, 10), test_rmse_errors, label = "TEST_RMSE")
plt.xlabel("MODEL COMPLEXITY")
plt.ylabel("ERRORS")
plt.legend() 
```
